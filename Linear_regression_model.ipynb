```python
import numpy as np
```

```python
#training set the values are multiples(x_train is 1000's and y_train is 1000's)
x_train = np.array([6.5,8.0,9.5,11.0,12.5,14.0,15.5,17.0,18.5,20.0])#features
y_train = np.array([140,165,185,210,235,255,280,305,325,350])#target values
```

```python
#computing cost function
def cost_function(x_train,y_train,w,b):
    m = x_train.shape[0]
    cost=0
    #defining the model
    for i in range(m):
      f_wb = w*x_train[i]+b
      cost = cost + (f_wb-y_train[i])**2
    total_cost = 1/(2*m)*cost
    return total_cost
```

```python
#computing gradient to update w and b simultaneously
def compute_gradient(x_train,y_train,w,b):
    m = x_train.shape[0]
    dj_dw = 0
    dj_db = 0
    for i in range(m):
      f_wb = w*x_train[i]+b
      dj_dw+=(f_wb-y_train[i])*x_train[i]
      dj_db+=(f_wb-y_train[i])
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_dw,dj_db
```

```python
def gradient_descent(x_train,y_train,w_in,b_in,alpha,num_iters,cost_function,compute_gradient):
  cost_history=[]
  params_history=[]
  w=w_in
  b=b_in
  for i in range(num_iters):
    dj_dw,dj_db = compute_gradient(x_train,y_train,w,b)
    #updating w and b values simulataneously
    b = b - alpha * dj_db
    w = w - alpha * dj_dw
    cost = cost_function(x_train,y_train,w,b)
    cost_history.append(cost)
    params_history.append([w,b])
    #Print for every 1000 iterations
    # print every 1000 iterations
    if i % 1000 == 0:
      print(f"Iteration {i}: Cost {cost:.2f}, w {w:.4f}, b {b:.4f}, dj_dw {dj_dw:.4f}, dj_db {dj_db:.4f}")
  return w,b,params_history,cost_history
```

```python
# initialize parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 10000
tmp_alpha = 1.0e-2
# run gradient descent
w_final, b_final, cost_history, params_history = gradient_descent(x_train ,y_train, w_init, b_init, tmp_alpha,
                                                    iterations, cost_function, compute_gradient)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```

```python
print(f"8000 sqft house prediction {w_final*8.0 + b_final:0.1f} Thousand dollars")
print(f"10000 sqft house prediction {w_final*10.0 + b_final:0.1f} Thousand dollars")
print(f"15000 sqft house prediction {w_final*15.0 + b_final:0.1f} Thousand dollars")
```
